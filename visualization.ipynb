**PIE CHART**

```python
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv("Iris.csv")
ax=plt.subplots(1,1,figsize=(10,8))
a['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(8,5))
plt.title("Iris Species ")
plt.show()
```
